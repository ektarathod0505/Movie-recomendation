```python
import ast

def convert(obj):
    obj = ast.literal_eval(obj)  # convert string to list of dicts
    L = []
    for i in obj:
        L.append(i['name'])
    return L

```

```python
import ast

def fetch_director(x):
    L = []
    try:
        data = ast.literal_eval(x)
        for i in data:
            if i.get('job') == 'Director':
                L.append(i['name'])
                break
    except:
        pass
    return L

```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
       
```

```python
def recommend(movie):
    # 1. Get the index of the selected movie
    movie_index = new_df[new_df['title'] == movie].index[0]

    # 2. Use that index to fetch similarity scores
    distances = similarity[movie_index]

    # 3. Sort the movies based on similarity (descending)
    movie_list = sorted(list(enumerate(distances)), key=lambda x: x[1], reverse=True)[1:6]

    # 4. Print recommended movie titles
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
